# Introduction Python

```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from scipy import stats
from sklearn.manifold import TSNE
import time
```

## Data Extraction

```python
stock_tesla = pd.read_csv('TSLA.csv')
tesla = stock_tesla.rename(columns = {'Open' : 'Tesla_Open'})
tesla.head()
```

```python
stock_facebook = pd.read_csv('FB.csv')
facebook = stock_facebook.rename(columns = {'Open' : 'Facebook_Open'})
facebook.head()
```

```python
stock_bbca = pd.read_csv('BBCA.JK.csv')
bbca = stock_bbca.rename(columns = {'Open' : 'BBCA_Open'})
bbca.head()
```

```python
stock_microsoft = pd.read_csv('MSFT.csv')
microsoft = stock_microsoft.rename(columns = {'Open' : 'Microsoft_Open'})
microsoft.head()
```

```python
stock_asii = pd.read_csv('ASII.JK.csv')
asii = stock_asii.rename(columns = {'Open' : 'ASII_Open'})
asii.head()
```

## Joining Table 

```python
join = [tesla['Tesla_Open'], facebook['Facebook_Open'], bbca['BBCA_Open'], microsoft['Microsoft_Open'], asii['ASII_Open']]
df = pd.concat(join, axis = True)
data_100 = df.iloc[0:101]
data_100
```

```python
join_date = [tesla['Date'], df]
all_stock = pd.concat(join_date, axis = True)
all_stock
```

```python
zscore_1 = stats.zscore(data_100, ddof = 1)
print(zscore_1)
```

```python
zscore_2 = stats.zscore(data_100, ddof = 2)
print(zscore_2)
```

## T-SNE Clustering Model

```python
n_iter = 1000
verbose = 1
n_components = 2
perplexities = [25, 30, 35]

for i, perplexity in enumerate(perplexities):

    t0 = time.time()
    tsne = TSNE(n_components = n_components, perplexity = perplexity,
                n_iter = n_iter, verbose = verbose).fit_transform(zscore_1, zscore_2)
    t1 = time.time()
    print("circles, perplexity = %d in %.2g sec" % (perplexity, t1 - t0))
    print('\n')
```

## Visualization

```python
plt.figure(figsize = (10,6))
plt.plot(zscore_1)
plt.title("Z Score with 1 Standard Deviation")
plt.legend(['Tesla', 'Facebook', 'BBCA', 'Microsoft', 'ASII'])
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(zscore_2)
plt.title("Z Score with 2 Standard Deviation")
plt.legend(['Tesla', 'Facebook', 'BBCA', 'Microsoft', 'ASII'])
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(tsne)
plt.title("T-SNE Clustering Plot Model")
plt.legend(['Z Score 1 Standard Deviation', 'Z Score 2 Standard Deviation'])
plt.show()
```
